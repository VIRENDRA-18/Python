```python
# Finonacci series
number = int(input("Enter number : "))
a = 0
b = 1
counter = 0
while counter < number :
    print(a)
    c = a + b
    a = b
    b = c
    counter = counter + 1
```

```python
word = input("Enter your string for reversal")
reverse = ""
length = len(word)
while length > 0 :
    reverse = reverse + word[length-1]
    length = length-1
print(reverse)    
```
