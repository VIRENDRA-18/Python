```python
import csv
```

```python
data = [['name', 'mail_id', 'phone_num'],
        ['Virendra', 'Virendrasingh203@gmail.com', 8384935677],
        ['Mudit', "muditsaini@gmail.com", 6346789790]
]
```

```python
with open('data.csv', 'w') as f :
    writer = csv.writer(f)

    for i in data :
        writer.writerow(i)
```

```python
with open('data.csv', 'r') as f :
    read_data = csv.reader(f)

    for i in read_data :
        print(i)
```
