List Of General Use Exceptions

```python
try :
    a = 10 / 0
except ZeroDivisionError as e :
    print(e)    
```

```python
try :
    int('viru')
except (ValueError, TypeError) as e :
    print(e)
```

```python
try :
    int('viru')
except :
    print('this will catch an error')    
```

```python
try :
    import viru
except ImportError as e :
    print(e)    
print('25 + 7 = ', 25 + 7)    
```

```python
try :
    d = {'key' : 'virendra', 1 : [1,2,3,4]}
    print(d['key2'])
except KeyError as e :
    print(e)    
```

```python
try :
    'viru'.test()
except AttributeError as e :
    print(e)    
```

```python
try : 
    l = [1,2,3,4,5]
    print(l[6])
except IndexError as e :
    print(e)    
```

```python
try :
    123 + 'vikram'
except TypeError as e :
    print(e)    
```

```python
try :
    with open ('exam.txt', 'r') as f :
        test = f.read()
except FileNotFoundError as e :
    print(e)        

    
```
