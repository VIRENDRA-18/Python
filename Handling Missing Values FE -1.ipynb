Imputation Missing Values

1- Mean Value Imputation

```python
sns.distplot(df['age'],kde=True)
```

```python
sns.histplot(df['age'],kde=True)
```

```python
df.age.isnull().sum()
```

```python
df['Age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['Age_mean','age']]
```

```python
## this tecnique work well when your data is normally distrubution
```

2 Median Value Imputation

if you have outliers in data set use this technique

```python
df['Age_median'] = df['age'].fillna(df['age'].median())
```

```python
df[['Age_median','Age_mean','age']]
```

3. Mode Imputation Technqiue--Categorical values

```python
df['embarked']
```

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df['age'].notna()
```

```python
df[df['age'].notna()]
```

```python
df[df['age'].notna()]['embarked'].mode()[0]
```

```python
mode = df[df['age'].notna()]['embarked'].mode()[0]
```

```python
mode
```

```python
df['embarked'].fillna(mode)
```

```python
df['embarked_mode']=df['embarked'].fillna(mode)
```

```python
df
```

```python
df[['embarked_mode','embarked']]
```
