```python
class pwskills1 :

    def students_detailed(self, name,email_id,number):
        print(name,email_id, number)

    @staticmethod
    def mentor_class(list_mentor):
        print(list_mentor)

    def mentor(self,mentor_list):
        print(mentor_list)    
            
```

```python
pwskills1.mentor_class(['krish', 'sudh'])
```

```python
std1 = pwskills1()
```

```python
std12 = pwskills1()
```

```python
std1.mentor(['viru','payel'])
```

```python
class pwskills2 :

    def students_detailed(self, name,email_id,number):
        print(name,email_id, number)

    @staticmethod
    def mentor_mail_id(mail_id_mentor):
        print(mail_id_mentor)    

    @staticmethod
    def mentor_class(list_mentor):
        pwskills2.mentor_mail_id(['sudh@gmail.com','krish@gmail.com'])
        print(list_mentor)

    @classmethod
    def class_name(cls):
        cls.mentor_class(['sudh','krish'])    

    def mentor(self,mentor_list):  # instance method , external method
        print(mentor_list)    
        self.mentor_class(['krish', 'viru'])

        ##  1. classmethod ke ander staticmethod ko kese use kre 
        ##   2. staticmethod ke ander staticmethod kese use kre 
        ##    3. instance method ke ander staticmethod ko kese add kre 
```

```python
pwskills2.class_name()
```

```python
pw = pwskills2()
```

```python
pw.mentor(['sudh','krish'])
```
