```python
import time

def timer_test(func) :
    def timer_test_inner():
        start = time.time()
        func()
        end = time.time()
        print(end-start)

    return timer_test_inner
```

Decorator ek special function hota hai jo kisi dusre function ya method ke behavior ko modify karta hai â€” bina uska original code change kiye.

```python
def test1() :
    print(45 + 10)
```

```python
test1()
```

```python
@timer_test
def test1() :
    print(45 + 10)
```
