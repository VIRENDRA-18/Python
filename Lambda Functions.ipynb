```python
n = 3
p = 2
```

```python
def test(n,p):
    return n**p
```

```python
test(3,2)
```

```python
# Lambda Function 

lambda n,p : n**p
```

```python
 
a = lambda n,p : n**p
```

```python
a(3,2)
```

```python
add = lambda x,y : x+y
add(4,10)
```

```python
c_to_f = lambda c : (9/5)*c + 32
c_to_f(47)
# Celsius to Fahrenheit conversion
```

```python
finding_max = lambda x,y : x if x>y else y
finding_max(90,67)
```
