```python
class pwskills1 :
    def __init__(self, name, email):
        self.name = name
        self.email = email
        
    @classmethod
    def detiled (cls, name, email)  :
        return cls(name,email)
      

    def students_detiled(self) :
        print(self.name, self.email)    
```

```python
pwskills1.detiled('praveen', 'jaat@gmail.com')
```

```python
pw1 = pwskills1.detiled('praveen', 'jaat@gmail.com')
```

```python
pw1.name
```

```python
pw1.email
```

```python
pw1.students_detiled()
```

```python
class pwskills2 :

    mobile_num = 8384567407

    def __init__(self, name, email):
        self.name = name
        self.email = email

    @classmethod
    def change_num (cls, mobile):                    ## overlanding concept bhi hai isme
        pwskills2.mobile_num = mobile    
        
    @classmethod
    def detiled (cls, name, email)  :
        return cls(name,email)
      

    def students_detiled(self) :
        print(self.name, self.email,pwskills2.mobile_num) 
```
