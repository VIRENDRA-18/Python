```python
def test_fib() :
    a,b = 0,1
    while True :
        yield a
        a,b = b, a+b
```

```python
fib = test_fib()
```

```python
for i in range(10):
    print(next(fib))
```

```python
def count_test(n) :
    count = 1
    while count < n :
        yield count
        count = count + 1
    
```

```python
c = count_test(5)
```

```python
for i in c:
    print(i)
```
