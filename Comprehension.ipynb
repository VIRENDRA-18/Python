```python
#Comprehension Way

[i ** 2 for i in list]
```

```python
# Ab mko Even Number chaiye list mai se

[ i for i in list if i % 2 == 0]
```

```python
# odd number chaye list mai se

[i for i in list if i % 2 != 0]
```
