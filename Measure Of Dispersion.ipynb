Measure Of Dispersion

```python
ages_lst = [23,43,23,56,74,32,68,98,45,32]
```

```python
import numpy as np
```

```python
np.mean(ages_lst)
```

```python
## varience

np.var(ages_lst)
```

```python
# std

np.std(ages_lst)
```

```python
import seaborn as sns
sns.histplot(ages_lst,kde=True)
```
